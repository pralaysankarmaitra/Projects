```python
from matplotlib.image import imread
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# upload and read your image here
image = imread('/content/image12.jpg')

# print image shape here

# plot the image
```

```python
# Understand what the following line does.
X = image.reshape(-1, 3)

# change and experiment with different number of clusters
kmeans = KMeans(n_clusters=4, random_state=101).fit(X)
```

```python
# print the cluster centers

# print the labels of each of row of X

# print the shape of the kmeans.labels_ array
```

```python
seg_img = []

for label in kmeans.labels_:
  seg_img.append(kmeans.cluster_centers_[label])

# print the contents and shape of seg_img after each of these lines
# and understand the working
seg_img = np.array(seg_img)
seg_img = seg_img.astype(int)
seg_img = seg_img.reshape(image.shape)
```

```python
plt.subplot(121)
plt.title('Original Image')
plt.imshow(image)

plt.subplot(122)
plt.title('Segmented 8 colors')
plt.imshow(seg_img)
plt.show()

# plot the segmented images for all the different cluster numbers you have tried
#
```
